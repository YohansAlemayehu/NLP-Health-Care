```python
frames = []
for i in range(1, 3):
    url = 'https://www.askapatient.com/viewrating.asp?drug=50670&name=ZITHROMAX&page={i}'
    payload = { 'api_key': '1bef27e7248bc0b02765e55b054a74a6', 'url': url, 'render': True, 'autoparse': True}
    r = requests.get('https://api.scraperapi.com/', params=payload)
    data = BeautifulSoup(r.content, 'html.parser')

    data_table = data.find_all('table')[1]
    table_titles = data_table.find_all('tr')[0]
    pt_table_titles = [title.text.strip() for  title in table_titles]
    df = pd.DataFrame(columns=pt_table_titles)
    col_data = data_table.find_all('tr')

    for row in col_data[2:]:
        rw_data = row.find_all('td')
        row_data = [data.text.strip() for data in rw_data]
        row_len = len(df)
        df.loc[row_len] = row_data
    frames.append(df)


    
    
```

```python
Data_df = pd.concat(frames)
Data_df
```
